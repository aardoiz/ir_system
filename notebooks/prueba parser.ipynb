```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
r = requests.get('https://ariesco.github.io/OIM/docs/intro.html')
soup = BeautifulSoup(r.text, 'html')
```

```python
soup
```

```python
soup.title
```

```python
soup.body
```

```python
body = str(soup.body)
```

```python
body
```

```python
body_noscript = re.sub('<script>(?:.+\n|\n)*<\/script>','',body)
body_notag = re.sub('<.+?>','', body_noscript)
body_clean = re.sub('\t', ' ', body_notag)
```

```python
body_clean
```

```python
print(body_clean)
```

```python
lista_parrafos = [n for n in body_clean.split('\n\n') if len(n) >0]
```
