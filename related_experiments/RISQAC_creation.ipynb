```python
from datasets import load_dataset
import pickle
import pandas as pd
```

```python
dataset = load_dataset("PlanTL-GOB-ES/SQAC")
```

```python
train = dataset['train']
test  = dataset['test']
val   = dataset['validation']
```

```python
#Ejemplo del tipo de datos que se trabajan
train[43]
```

```python
count = 0
nombres = []
data=[]
for element in train:
    if element['title'] in nombres:
        count += 1
    else:
        nombres.append(element['title'])
        dicto = {}
        dicto['title'] = element['title']
        dicto['content'] = element['context']
        dicto['question'] = element['question']
        data.append(dicto)

for element in test:
    if element['title'] in nombres:
        count += 1
    else:
        nombres.append(element['title'])
        dicto = {}
        dicto['title'] = element['title']
        dicto['content'] = element['context']
        dicto['question'] = element['question']
        data.append(dicto)

for element in val:
    if element['title'] in nombres:
        count += 1
    else:
        nombres.append(element['title'])
        dicto = {}
        dicto['title'] = element['title']
        dicto['content'] = element['context']
        dicto['question'] = element['question']
        data.append(dicto)
```

# Sacamos las preguntas y las agrupamos por título

```python
count = 0
preguntas={}
for element in train:
    if element['title'] in preguntas.keys():
        preguntas[element['title']].append(element['question'])
    else:
        preguntas[element['title']] = [element['question']]

for element in test:
    if element['title'] in preguntas.keys():
        preguntas[element['title']].append(element['question'])
    else:
        preguntas[element['title']] = [element['question']]        

for element in val:
    if element['title'] in preguntas.keys():
        preguntas[element['title']].append(element['question'])
    else:
        preguntas[element['title']] = [element['question']]        
```
